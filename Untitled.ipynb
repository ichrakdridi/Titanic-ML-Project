WEEK 2 : Preprocessing & Feature Engineering : Objectif : préparer les données pour le modèle ML (imputation, encodage, scaling, nouvelles features).

```python
# Imputation des valeurs manquantes
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

# Vérification
df.isnull().sum()

```

```python
# Encodage des colonnes catégorielles

# Sex : Male=0, Female=1
df['sex'] = df['sex'].map({'male': 0, 'female': 1})

# Embarked : One-Hot Encoding
df = pd.get_dummies(df, columns=['embarked'], drop_first=True)

# Bool → int
df['alone'] = df['alone'].astype(int)
df['adult_male'] = df['adult_male'].astype(int)

df.head()

```

Cellule 3 : Feature Engineering

```python
# Création de nouvelles features
# Taille de la famille
df['family_size'] = df['sibsp'] + df['parch'] + 1

df.head()
```

Cellule 4 : Séparer features et target

```python
# Séparer X et y
X = df.drop(columns=['survived'])
y = df['survived']

X.head()
y.head()
```

Cellule 5 : Split train/test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=2025
)

X_train.shape, X_test.shape
```

Cellule 6 : Pipelines pour preprocessing

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.impute import KNNImputer
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier

# Colonnes numériques et catégorielles
numerical_columns = X.select_dtypes(include=['float', 'int']).columns
categorical_columns = X.select_dtypes(include=['object']).columns

# Pipeline numérique : imputation + scaling
numerical_processing = Pipeline([
    ('imputer', KNNImputer(n_neighbors=5)),
    ('scaler', MinMaxScaler())
])

# Pipeline catégoriel : imputation + one-hot
categorical_processing = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder())
])

# ColumnTransformer
preprocessing = ColumnTransformer([
    ('num', numerical_processing, numerical_columns),
    ('cat', categorical_processing, categorical_columns)
])
```

Cellule 7 : Pipeline complet et entraînement

```python
# Pipeline complet avec RandomForest
pipe = Pipeline([
    ('preprocessing', preprocessing),
    ('model', RandomForestClassifier(random_state=2025))
])

pipe.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

y_pred = pipe.predict(X_test)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Rapport de classification
print(classification_report(y_test, y_pred))

# Matrice de confusion
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
