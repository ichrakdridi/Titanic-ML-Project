```python
data.drop(columns=['deck'], inplace= True)
```

```python
# data['family_size'] = data.sibsp + data.parch +1
```

```python
x = data.drop(columns=['survived'])
y = data['survived']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.2,  
                                                    stratify= y,random_state=2025)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier

numerical_columns = x.select_dtypes(include = ['float', 'int']).columns
categorical_columns = x.select_dtypes(include = ['object']).columns
```

```python
# Imputation and scaling
numerical_processing = Pipeline(
    [
        ('imputer', SimpleImputer(strategy='median')),
        ('scaling', MinMaxScaler())
    ]
)

categorical_processing = Pipeline(
    [
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder())
    ]
)
```

```python
preprocessing = ColumnTransformer(
    transformers = [
        ('num', numerical_processing, numerical_columns  ),
        ('cat',categorical_processing,categorical_columns  ),
    ]
)
```

```python
pipe = Pipeline(
    [
        ('preprocessing', preprocessing),
        ('model', RandomForestClassifier(random_state=2025))
    ]
)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
accuracy_score(y_test, y_pred)
#pipe.score(x_test, y_test)
```

```python
params = {
   'model__n_estimators' : [20, 40, 60, 80, 100],
   'model__max_depth' : [3, 4, 6, 10, None],
   'model__min_samples_split' : [3, 4, 6,8],
   'model__min_samples_leaf': [3,4, 6, 8]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
search = GridSearchCV(pipe, param_grid=params, cv=4)
```

```python
search.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = search.predict(x_test)
accuracy_score(y_test, y_pred)
```
